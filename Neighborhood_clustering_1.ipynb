```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
html_content = pd.read_html(url, header = 0)
df = html_content[0]
```

# Checking whether Borough is Not assigned in the dataframe

```python
df[(df.Borough == 'Not assigned')]
```

```python
# Remove dataframw where Borough is not assigned
```

```python
df = df[(df.Borough != 'Not assigned')]
```

```python
df.head()
```

# Group the Neighborhoo which has same postal code and borough

```python
temp_value=df.groupby(['Postal code','Borough'])['Neighborhood'].transform(lambda x:','.join(x))
```

```python
df.loc[:,'Neighborhood']=temp_value.copy(deep=True)
```

```python
df.head()
```

```python
df = df.drop_duplicates()
df.reset_index(drop = True, inplace=True)
```

```python
df.head(10)
```

# Assign neighborhood as borough if nwighborhood is not assigned 

```python
df.Neighborhood[df['Neighborhood']=='Not assigned'] = df.Borough[df['Neighborhood']=='Not assigned']
```

```python
def replace_string(str):
    return str.replace('/',',')
```

```python
df.Neighborhood=df.Neighborhood.map(replace_string)
```
